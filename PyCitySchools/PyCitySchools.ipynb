```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# school_data.head()
# student_data.head()
# school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the data requested
total_schools = school_data["School ID"].count()
total_students = student_data["Student ID"].count()
total_budget = school_data["budget"].sum()
average_math_score = student_data["math_score"].mean()
average_read_score = student_data["reading_score"].mean()
passing_math = student_data.loc[student_data["math_score"]>= 70, "Student ID"].count() / total_students
passing_read = student_data.loc[student_data["reading_score"]>= 70, "Student ID"].count() / total_students
passing_overall = (average_math_score + average_read_score)/2

# Create a dataframe for the data
# intialise data of lists
district_summary = {
    "Total Schools" : total_schools,
    "Total Students" : total_students,
    "Total Budget" : total_budget,
    "Average Math Score" : average_math_score,
    "Average Reading Score" : average_read_score,
    "% Passing Math" : passing_math,
    "% Passing Reading" : passing_read,
    "% Overall Passing Rate" : passing_overall
}


# Create DataFrame 

district_summary_df = pd.DataFrame(district_summary, index=[0])

district_summary_df = district_summary_df.style.format({
    "Total Schools": '{:,.0f}'.format,
    "Total Students": '{:,.0f}'.format,
    "Total Budget": '${:,.0f}'.format,
    "Average Math Score": '{:,.4f}'.format,
    "Average Reading Score": '{:,.4f}'.format,
    "% Passing Math": '{:.4%}'.format, 
    "% Passing Reading": '{:.4%}'.format,
    "% Overall Passing Rate" : '{:,.4f}'.format   
})


district_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.

```python
df_9th = school_data_complete[school_data_complete["grade"] == "9th"]
df_9th_grouped = df_9th.groupby(['school_name'])
mean_math_9th = df_9th_grouped["math_score"].mean()
mean_read_9th = df_9th_grouped["reading_score"].mean()

df_10th = school_data_complete[school_data_complete["grade"] == "10th"]
df_10th_grouped = df_10th.groupby(['school_name'])
mean_math_10th = df_10th_grouped["math_score"].mean()
mean_read_10th = df_10th_grouped["reading_score"].mean()

df_11th = school_data_complete[school_data_complete["grade"] == "11th"]
df_11th_grouped = df_11th.groupby(['school_name'])
mean_math_11th = df_11th_grouped["math_score"].mean()
mean_read_11th = df_11th_grouped["reading_score"].mean()

df_12th = school_data_complete[school_data_complete["grade"] == "12th"]
df_12th_grouped = df_12th.groupby(['school_name'])
mean_math_12th = df_12th_grouped["math_score"].mean()
mean_read_12th = df_12th_grouped["reading_score"].mean()


# Creating a new DataFrame for math
math_by_grad = pd.DataFrame({
                                "9th": mean_math_9th,
                                "10th": mean_math_10th,
                                "11th": mean_math_11th,
                                "12th" : mean_math_12th,
                            })

# Creating a new DataFrame for reading
reading_by_grad = pd.DataFrame({
                                "9th": mean_read_9th,
                                "10th": mean_read_10th,
                                "11th": mean_read_11th,
                                "12th" : mean_read_12th,
                            })
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
# school_data_complete.columns
```

```python
school_data_complete["Spending Per Student"] = school_data_complete["budget"] / school_data_complete["size"]

school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["Spending Per Student"], spending_bins, labels=group_names)
                                  

scores_by_spending = school_data_complete.groupby(["Spending Ranges (Per Student)", "Spending Per Student"])




total_students = scores_by_spending["Student ID"].count()

total_school_budget = scores_by_spending["budget"].mean()
budget_student = total_school_budget / total_students

average_math_score = scores_by_spending["math_score"].mean()
average_read_score = scores_by_spending["reading_score"].mean()
passing_math = scores_by_spending["passing_math"].sum()/ total_students
passing_reading = scores_by_spending["passing_read"].sum()/ total_students
overall_passing = (scores_by_spending["passing_math"].sum() + scores_by_spending["passing_read"].sum()) / total_students / 2 



# Creating a new DataFrame 
scores_by_spending_summary = pd.DataFrame({
                                    "Average Math Score" : average_math_score,
                                    "Average Reading Score" : average_read_score,
                                    "% Passing Math" : passing_math,
                                    "% Passing Reading" : passing_reading,
                                    "% Overall Passing Rate" : overall_passing
                                   })

# Used to calculate the mean of means
scores_by_spending_summary = scores_by_spending_summary.groupby(["Spending Ranges (Per Student)"]).mean()

# Formats the dataframe
scores_by_spending_summary = scores_by_spending_summary.style.format({
                                    "Average Math Score" : '{:,.4f}'.format,
                                    "Average Reading Score" : '{:,.4f}'.format,
                                    "% Passing Math" : '{:,.4%}'.format,
                                    "% Passing Reading" : '{:,.4%}'.format,
                                    "% Overall Passing Rate" : '{:,.4%}'.format
                                    })


scores_by_spending_summary 

```
